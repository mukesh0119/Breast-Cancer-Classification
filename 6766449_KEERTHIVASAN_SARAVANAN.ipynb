# Preprocessing :
I must ensure that the data has undergone all necessary preprocessing before modelling it. I divided the data into a train set and a test set during preprocessing in order to prevent overfitting. I must first divide the data into two parts, one of which is filled with input attributes and the other with the targeted attribute. Here, I've used Class as the target attribute and all the other attributes as the input data.

```python
x = dsp_data.drop(["Class"], axis = 1)
y = dsp_data.Class.values
```

```python
x.head()
```

```python
y[:10]
```

### Splitting into train and test data : 
Before implementing the Naive Bayes and Logistic Regression,the data must first be divided into training and testing sets. I divided the dataset into a train set and a test set even When I was only dealing with one data set. To avoid overfitting, the dataset is split into two parts. The model will be trained using the train set, and predictions will subsequently be made using the test set. I divided the dataset using Sklearn's train-test split method. I divided the data, 70% for train set and 30% for test.

```python
x_tr, x_te, y_tr, y_te = st(x, y, test_size = 0.3, random_state = 42)
```

### Scaling the data : 
Our data contains numerical values. I scaled the data because I don't know the units for each attribute and modelling the data with varied unit ranges will lead to bad model performance. In this instance, I scaled the data using the MinMax scaler. Using this scaler it scales the data from the range of -1 to +1. The formula for this scaler is : 

m = (x -xmin) / (xmax -xmin)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sx_tr=x_tr
for col in sx_tr:
    z=MinMaxScaler()
    sx_tr[col]=z.fit_transform(sx_tr[[col]])
```

```python
sx_te=x_te
for col in sx_te:
    z=MinMaxScaler()
    sx_te[col]=z.fit_transform(sx_te[[col]])
```

# Naive bayes : 
Gaussian Naive Bayes is a strong machine learning method used for classifying problems. It is a simple and fast method for creating prediction models. The basis of this method is that each feature in data is independent to the others. To calculate the chances of a specific result, probability is used. I'll be using the Gaussian Naive Bayes technique to forecast the results of our data. The Scikit-Learn GaussianNB() library will be used for impleting the model. As input data for our dataset, I've used all the  attributes as input. Even though I know from the corrplot that "Cell.size," "Cell.shape," and "Bare.nuclei" are strongly correlated to the targeted "Class", I'm still using all attributes as input since it will improve the classification. There are three types of naive bayes classifiers. The reason I'm using gaussianNB() is our dataset is descrete form.


Naive Bayes mainly uses the Bayes rule to perform classification. The formula for Bayes rule is:
P (A | B) = P (A âˆ© B) / P(B)

```python
bayes = GaussianNB();
bayes.fit(sx_tr, y_tr);
```

```python
y_p = bayes.predict(sx_te)
v.hist(y_p,edgecolor='black');
```

### Accuracy of Naive Bayes :
In order to see whether the Naive Bayes model is working as expected, it is mandatory to assess accuracy of the model after fitting it with the datas. The percentage of accurately predicted values to the total number of samples is what makes up the accuracy score. By using the.score() function from the sklearn library on the trained model,we can determine the score.

```python
print("Train data accuracy: ",bayes.score(sx_tr, y_tr))
print("Test data accuracy: ",bayes.score(sx_te, y_te))
```

### Classification report : 
To calculate the performance of naive bayes model we have to find out the precision, accuracy and f1 score. I'm using the classification report from  Sklearn lib to calculate these results.

```python
from sklearn.metrics import classification_report as cr
print(cr(y_te, y_p))
```

# Logistic Regression : 
Logistic Regression is an algorithm for creating predictions based on a collection of input variables.
To implement the logistic regression model, we will use the Scikit-learn library. In Logistic Regression prediction of a binary output is done using the input data. In the given dataset the targeted attribute that we need to predict by training the logistic regression model is the column "Class". As input data for our dataset, I used all the attributes. The targeted column is 'Class'. As i mentioned before, I'll take all attributes as input data for the better perfomance of the model.

The formula for Logistic Regression is : logit(p_{X}) = log(p_{X}/(1 - p_{X})) = beta_{0} + beta_{1}*X

```python
log= LogisticRegression()
log.fit(sx_tr, y_tr);
```

```python
y_p1 = log.predict(sx_te)
v.hist(y_p1);
```

### Accuracy of Logistic Regression :

### Comparing the performance : 
To determine how well the models performed, I utilised the precision score, f1 score, and recall score functions from the Sklearn library. I've assigned these numbers to a dataframe after discovering all of these values so that I can visualise them. The formula for calculating the metrics are :

Precision = TruePositives / (TruePositives + FalsePositives)

Recall = TruePositives / (TruePositives + FalseNegatives)

F1= (2 * Precision * Recall) / (Precision + Recall)

```python
comp = p.DataFrame()

comp['accuracy'] = [ bayes.score(sx_te, y_te),log.score(sx_te, y_te)]
comp['precision'] = [ps(y_te, y_p),ps(y_te, y_p1)]
comp['recall score'] = [rs(y_te, y_p),rs(y_te, y_p1)]
comp['f1 score'] = [f1(y_te, y_p),f1(y_te, y_p1)]
```

```python
comp1=comp.T
comp1.columns = ['Naive_Bayes','Logistic_Regression']
```

### Learning Curves : 
A learning curve shows how well a model learns.The model can be tested on the training dataset plus a validation dataset following each update during training.If maximisation is the metric used to measure learning, then higher scores signify more learning.

```python
from sklearn.model_selection import learning_curve
```

### Logistic reg :

```python
trsz, trsc, tesc = learning_curve(LogisticRegression(), x_te, y_te,
                                                        cv=50,
                                                        train_sizes=n.linspace(0.01, 1.0, 100))
```

```python
trme = n.mean(trsc, axis=1)
trsd = n.std(trsc, axis=1)
teme = n.mean(tesc, axis=1)
tesd = n.std(tesc, axis=1)
```

```python
v.subplots(figsize=(20,10))
v.plot(trsz, trme, '--',  label="Training score")
v.plot(trsz, teme, label="Cross-validation score")


v.title("LogisticRegression Learning Curve")
v.xlabel("Training Size"), v.ylabel("Accuracy"),v.legend(loc="best");
```

### Naive Bayes :

```python
trsz, trsc, tesc = learning_curve(GaussianNB(), x_te, y_te,
                                                        cv=50,
                                                        train_sizes=n.linspace(0.01, 1.0, 100));
```

```python
trme = n.mean(trsc, axis=1)
trsd = n.std(trsc, axis=1)
teme = n.mean(tesc, axis=1)
tesd = n.std(tesc, axis=1)
```

```python
v.subplots(figsize=(20,10))
v.plot(trsz, trme, '--',  label="Training score")
v.plot(trsz, teme, label="Cross-validation score")
v.title("Naive bayes Learning Curve")
v.xlabel("Training Size"), v.ylabel("Accuracy"),v.legend(loc="best");
```
